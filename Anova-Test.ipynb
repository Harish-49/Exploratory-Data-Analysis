### Finding Hypothesis:

Here the dataset denotes the salse rates of different regions, we need to increase the salse rate for that assuming Hypothesis.

 - Used to compare means of two or more samples, to check whether the means are similar or not
 - For Example - Sales = Anova can be used to compare mean of Sales rates of different regions whether to increase their customization or not to increase the Salse rate.

 - NULL HYPOTHESIS = means of two or more groups are similar to each other
 - Alternate Hypothesis = means of two or more grups are different than each other.

 For a CI = 95%, alpha = 1 - CI = 1 - 0.95 = 0.05

 - if pvalue > alpha = accept the Null Hypothesis, feature is not important
 - if pvalue < alpha = reject the Null Hypothesis, feature is important

### Creating Data frames

```python
d=pd.DataFrame(df,columns=['Region','Sales'])
d
```

```python
s=d[d['Region']=='South']
s
```

```python
e=d[d['Region']=='East']
e
```

```python
w=d[d['Region']=='West']
w
```

### Applying Anavo test

```python
from scipy import stats
```

```python
f,p=stats.f_oneway(s['Sales'],e['Sales'],w['Sales'],c['Sales'])
print(f)
print(p)
```

```python
cf=95/100
cf
```

```python
al=1-cf
al
```

```python
if p>al:
    print("NH Accept")
    print("Need not to costomize")
else:
    print("NH Reject")
    print("Need to costomize")
```

Here NH means Null Hypothesis, we reject NH so we need not to customise the Store.
